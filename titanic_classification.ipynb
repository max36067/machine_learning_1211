```python
import pandas as pd
train_df = pd.read_csv('titanic/train.csv',encoding='utf-8')
test_df = pd.read_csv('titanic/test.csv',encoding='utf-8')
```

```python
train_df
```

```python
test_df
```

```python
# 先把答案剝出來
y_train = train_df['Survived']
```

```python
x_train = train_df.drop(['Survived'], axis=1)
x_test = test_df
x_train
```

```python
# concat 為合併兩個表格, data的部分要使用 list框起來
# ignore_index 可以把索引數字重新計算
total = pd.concat([x_train,x_test],ignore_index=True)
total
```

```python
# 確定有哪些欄位是有缺值
total.isna().sum()
```

```python
# 補數字類型缺失值
med = total.median()
med
```

```python
total = total.fillna(med)
total.isna().sum()
```

```python
most = total['Embarked'].value_counts().idxmax()
total['Embarked'] = total['Embarked'].fillna(most)
total.isna().sum()
```

```python
# 整理一下
total = total.drop(['Cabin','PassengerId','Ticket'],axis=1)
```

```python
def namecut(n):
    return n.split(',')[-1].split('.')[0].replace(' ','')
total['Name'].apply(namecut).value_counts()
```

```python
def namecut(n):
    mid = n.split(',')[-1].split('.')[0].replace(' ','')
    if mid =='Mr' or mid == 'Sir':
        return 'Mr'
    elif mid == 'Mrs' or mid == 'Mme' or mid == 'Lady':
        return 'Mrs'
    elif mid == 'Miss' or mid == 'Mlle':
        return 'Miss'
    elif mid == 'Master':
        return 'Master'
    else:
        return None
total['Name'] = total['Name'].apply(namecut)
```

```python
# One Hot Encoding方法
total = pd.get_dummies(total)
total
```

```python
# iloc為拿出行的方法
x_train_process = total.iloc[:train_df.shape[0]]
x_test_process = total.iloc[train_df.shape[0]:]
x_test_process
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.figure(figsize=(14,14))
temp = pd.concat([y_train,x_train_process], axis=1)
# corr 為正負相關係數,畫圖出來看是否與生存有相關
sns.heatmap(temp.corr(),
            annot=True,
            cmap="RdBu_r")
```
